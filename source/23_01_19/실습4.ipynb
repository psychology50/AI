```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

fish_data = np.column_stack((fish_length, fish_weight)) # column을 기준으로 병합
fish_target = np.concatenate((np.ones(35), np.zeros(14))) # one-hot coding으로 분류

# 훈련/테스트 세트 분리
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, random_state=42)

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
kn.predict([[25, 150]])

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^') # 새로운 도미 추가해보기
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

distances, indexes = kn.kneighbors([[25, 150]]) # 최근접 알고리즘이 구한 정점과의 거리와 인덱스 반환

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show() # 도미보다 빙어가 최단거리 매칭이 많음. x,y 축의 차이로 인한 최단거리 오류

# 명확하게 확인하기
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlim((0, 1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show() # x축의 값의 변화를 줬을 때, y축에 일렬로 가깝게 세우면 빙어와 가까움

# 기준 맞추기
# 스케일 통일 기법: 표준 점수 = (특성 - 평균)/표준편차 -> 평군에서 표준편차의 몇 배만큼 떨어져 있는가?
# 훈련/테스트 세트 모두 스케일 조정을 해야 한다.
# 단, 분류 문제에서는 필수지만 회귀 문제에서는 스케일 과정이 필요 없다. (ex, 트리에서는 스케일 조정 작업 불필요)

mean = np.mean(train_input, axis=0) # 평균
std = np.std(train_input, axis=0)   # 표준 편차
print(mean, std)

train_scaled = (train_input - mean) / std

# 전처리 데이터로 모델 훈련
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show() # 새로운 도미를 스케일 변환 없이 스케일 변환된 훈련 데이터와 그리는 경우 오류

new = ([25,150] - mean) / std # new 데이터도 스케일 변환 시작

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^') # new[0]: 길이, new[1]: 무게
plt.xlabel('length')
plt.ylabel('weight')
plt.show() # x, y축 기준값 유의

kn.fit(train_scaled, train_target) # 학습 시작
test_scaled = (test_input - mean) / std
kn.score(test_scaled, test_target)

print(kn.predict([new])) # [1.]

distances, indexes = kn.kneighbors([new])

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes,0], train_scaled[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression


perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 
     1000.0, 1000.0]
     )

# 훈련 세트와 테스트 세트로 나눕니다
train_input, test_input, train_target, test_target = train_test_split(
    perch_length, perch_weight, random_state=42)
# 훈련 세트와 테스트 세트를 2차원 배열로 바꿉니다
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)

knr = KNeighborsRegressor(n_neighbors=3)
# k-최근접 이웃 회귀 모델을 훈련합니다
knr.fit(train_input, train_target)


# 50cm 농어의 이웃을 구합니다
distances, indexes = knr.kneighbors([[50]])

# 훈련 세트의 산점도를 그립니다
plt.scatter(train_input, train_target)
# 훈련 세트 중에서 이웃 샘플만 다시 그립니다
plt.scatter(train_input[indexes], train_target[indexes], marker='D')
# 50cm 농어 데이터
plt.scatter(50, 1033, marker='^')
plt.show() 


# 100cm 농어의 이웃을 구합니다
distances, indexes = knr.kneighbors([[100]])

# 훈련 세트의 산점도를 그립니다
plt.scatter(train_input, train_target)
# 훈련 세트 중에서 이웃 샘플만 다시 그립니다
plt.scatter(train_input[indexes], train_target[indexes], marker='D')
# 100cm 농어 데이터
plt.scatter(100, 1033, marker='^')
plt.show() # 100cm가 들어와도 무조건 농어가 된다..알고리즘이 학습하도록 만들자.


# 선형 회귀 시작 (샘플 범위 밖의 값을 극복)
# 특성이 하나인 경우, 분류 기준을 찾는 어떤 직선을 머신 러닝이 직접 학습하는 것
lr = LinearRegression()
lr.fit(train_input, train_target) # 직선 방정식을 구하는 과정

# 50cm 농어에 대한 예측 (인공지능이 학습한 결과)
print(lr.predict([[50]])) # [1241.83860323] (50cm -> 1.033kg에서 1.24kg 정확도 향상)
print(lr.coef_, lr.intercept_) # [39.01714496] -709.0186449535477

# 추세선 그리기
plt.scatter(train_input, train_target)
plt.plot([15, 50], [15*lr.coef_ + lr.intercept_, 50*lr.coef_ + lr.intercept_])
plt.scatter(50, 1241.8, marker='^')
plt.show()

# 하지만 여전히 학습도가 떨어진다.
print(lr.score(train_input, train_target)) # 0.939846333997604
print(lr.score(test_input, test_target))   # 0.8247503123313558


# 다항 회귀 시작
# 다중 회귀와는 다르다.
# 최적의 직선을 찾기보다는 최적의 곡선을 찾자.
# 무게 = a*길이^2 + b*길이 + c
train_poly = np.column_stack((train_input**2, train_input))
test_poly = np.column_stack((test_input**2, test_input))

lr = LinearRegression()
lr.fit(train_poly, train_target)

print(lr.predict([[50**2, 50]])) # [1573.98423528] (다항 입력 형태로 예측값 준비, 1.241kg -> 1.573kg 정도 예측)
print(lr.coef_, lr.intercept_)   # [  1.01433211 -21.55792498] 116.0502107827827 (제곱항이 추가되어 계수가 2개)
# 새로운 다항식 -> 무게 = 1.01*길이^2 - 21.6*길이 + 116.05

# 구간별 직선을 그리기 위해 15에서 49까지 정수 배열을 만듭니다
point = np.arange(15, 50) # 짧은 직선을 이어서 곡선으로 표기
plt.scatter(train_input, train_target)
# 15에서 49까지 2차 방정식 그래프를 그립니다
plt.plot(point, 1.01*point**2 - 21.6*point + 116.05)
plt.scatter([50], [1574], marker='^')
plt.show()

print(lr.score(train_poly, train_target)) # 0.9706807451768623
print(lr.score(test_poly, test_target))   # 0.9775935108325122
# 테스트 점수가 조금 더 높다 (과소적합이 여전히 존재한다.) -> 추가 개선이 필요하다.
```
